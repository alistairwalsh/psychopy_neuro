- It can access hardware and internet resources, local or remote databases and many different forms of input as it is built on python

------------
Many experiments are based around the idea of testing a response at various levels of the same or a small number of variables. Psychopy takes up this idea with routines and loops. A routine is the generic scenario and the loop repeats this basic routine with a change in the variable of interest.  

This can be done in a range between values or Psychopy can read data from a comma seperated value (CSV) spreadsheet. This spreadsheet can be used to specify what stimuli should be presented, the correct answer, degrees of rotation, number of distractors, opacity, etc. Variables to change can be randomly generated or can be preset with the spreadsheet.
